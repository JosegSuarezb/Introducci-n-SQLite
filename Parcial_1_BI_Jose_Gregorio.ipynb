Preguntas para Restaurantes
1. ¿Que restaurantes hacen comida italiana
2. ¿Cuales restaurantes hacen tipo de comida saludable?
3. ¿Cuales restaurantes y que servicio ofrecen los que se encuentran en medellin?
4. ¿Cuales restaurantes solo hacen domicilio?

```python
#1. ¿Que restaurantes hacen comida italiana?
consulta = ''' SELECT Restaurante
              FROM "Tipo de comida"
              WHERE "Tipo de comida" = "Italiana"
              '''  #en este le pedimos que nos seleccion los restaurantes de la hoja de tipo de comida donde el tipo de comida sea italiana
cursor.execute(consulta)
Restaurantes = cursor.fetchall()
print(Restaurantes)
```

```python
#2. ¿Cuales restaurantes hacen tipo de comida saludable?
consulta = ''' SELECT Restaurante
              FROM "Tipo de comida"
              WHERE "Tipo de comida" = "Saludable" ''' #en el codigo le pedimos que nos seleccione los restaurantes de la hoja de tipo de comida donde su tipo sea saludable
cursor.execute(consulta)
Restaurantes_saludables = cursor.fetchall()
print(Restaurantes_saludables)
```

```python
#3 ¿Cuales restaurantes y que servicio ofrecen los que se encuentran en medellin?
consulta = ''' SELECT Restaurante,
                      Servicio
                  FROM "Restaurante tipo de comida"
                  WHERE Ubicación = "Medellin" ''' #aca le pedimos que nos seleccione los restaurantes y que servicio ofrecen de la hija de restaurante tipo de comida el cual la ubicacion sea medellin
cursor.execute(consulta)
Restaurantes_medellin=cursor.fetchall()
print(Restaurantes_medellin)
```

```python
#4. ¿Cuales restaurantes solo hacen domicilios?
consulta = ''' SELECT Restaurante
              FROM "Restaurante tipo de comida"
              WHERE Servicio = "Domicilio" ''' #le pedimos que nos los restaurantes de la hoja de restaurante tipo de comida en el que su servicio sea solo domicilio
cursor.execute(consulta)
Restaurantes = cursor.fetchall()
print(Restaurantes)
```

Preguntas para torneos
1. ¿Cuales deportes se practican en grupos?
2. ¿Que torneos son de categoria profesional?
3. ¿Cuales deportes y sus nombre de los torneos que se juegan en Estados Unidos?
4. ¿que torneos se juegan individualmente?

```python
#1. ¿Cuales deportes se practican en grupos?
Consulta = ''' SELECT Deporte
              FROM Torneos
              WHERE "Tipo de deporte" = "Individual" ''' #Le pedimos que nos seleccione los deportes de la hoja de torneos que su tipo sea solo individual
cursor.execute(Consulta)
Deporte=cursor.fetchall()
print(Deporte)
```

```python
#2. ¿Que torneos son de categoria profesional?
Consulta = ''' SELECT "Nombre del Torneo"
              FROM "Categoria de los torneos"
              WHERE Categoria = "Profesional" ''' #le pedimos que nos diga el nombre de los torneos de la hoja de Categoria de los torneos que se juegue profesionalmente
cursor.execute(Consulta)
Torneos=cursor.fetchall()
print(Torneos)
```

```python
#3. ¿Cuales deportes y sus nombre de los torneos que se juegan en Estados Unidos?
Consulta = ''' SELECT Deporte,
                     "Nombre del Torneo"
                     FROM Torneos
                     WHERE Ubicación = "Estados Unidos" ''' #le pedimos que nos seleccione los deportes y los torneos de la hoja de torneos donde se jueguen en Estados Unidos
cursor.execute(Consulta)
Deporte=cursor.fetchall()
print(Deporte)
```

```python
#4. ¿que torneos y donde se encuentran estos que se juegan individualmente?
Consulta = ''' SELECT "Nombre del Torneo",
                     Ubicación
                     FROM Torneos
                     WHERE "Tipo de deporte" = "Individual" ''' #le pido que me de los nombres de los torneos y donde se hacen de la hoja de torneos y que se jueguen individualmente
cursor.execute(Consulta)
Torneos=cursor.fetchall()
print(Torneos)
```
